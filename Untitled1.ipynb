```python
import os
import glob
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
import cv2

import tensorflow as tf
import tensorflow.keras as keras
from keras.layers import Dense, Dropout, Activation, Flatten, Input
from keras.optimizers import SGD
from keras import models, optimizers, layers
from keras.models import Model, Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from keras.preprocessing.image import array_to_img, img_to_array, load_img

from keras import backend as K 
```

```python
# クラス数
num_class = 2

# データ格納先（クラスごとに分けられたフォルダがあること）
dir = "/content/drive/MyDrive/image_demo2"

# 画像のリサイズ
image_size = 224

# 水増し
aug = False

# ステップ数の増加率（水増し時）
steps_rate = 1

# エポック数
epochs = 150

# バッチサイズ
batch_size = 32

# 使用モデル
BASE_MODEL='VGG16'

# 転移学習
transfer_learning = False

# 最適化手法
optim = "SGD"

# 早期終了
early_stop = True
patience = 10

# 学習率の更新
# 評価値の改善が止まった時に学習率を減らします．
reduce_rl = False

# 重みの保存
checkpoint = False
```

```python
classes = os.listdir(dir)
if len(classes) != num_class:
    num_class = len(classes)

input_shape = (image_size, image_size, 3)

def preprocess(dir):
    x = []
    y = []

    for idx, cls in enumerate(classes):
        print(cls)
        file_lists = glob.glob(os.path.join(dir, classes[idx]) + "/*.png")
        for file in file_lists:
            image = Image.open(file)
            image = image.convert("RGB")    
            image = image.resize((image_size, image_size))
            data = np.asarray(image)
            x.append(data)
            y.append(idx)

    x = np.array(x)
    y = np.array(y)

    x = x.astype('float32')
    x /= 255

    y = keras.utils.to_categorical(y, num_class)
    return x, y

x, y = preprocess(dir)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)

steps_per_epoch = np.ceil(len(x_train)/batch_size * steps_rate)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range = 0.2,
    channel_shift_range = 0.2,
    horizontal_flip=True,
)
datagen.fit(x_train)
```

```python
if BASE_MODEL=='VGG16':
    from keras.applications.vgg16 import VGG16 as PTModel
elif BASE_MODEL=='ResNet50V2':
    from keras.applications.resnet_v2 import ResNet101V2 as PTModel
elif BASE_MODEL=='InceptionV3':
    from keras.applications.inception_v3 import InceptionV3 as PTModel
elif BASE_MODEL=='InceptionResNetV2':
    from keras.applications.inception_resnet_v2 import InceptionResNetV2 as PTModel
elif BASE_MODEL=='Xception':
    from keras.applications.xception import Xception as PTModel
elif BASE_MODEL=='NASNetLarge': 
    from keras.applications.nasnet import NASNetLarge as PTModel
elif BASE_MODEL=='DenseNet121':
    from keras.applications.densenet import DenseNet121 as PTModel
elif BASE_MODEL=='DenseNet201':
    from keras.applications.densenet import DenseNet201 as PTModel
elif BASE_MODEL=='MobileNetV2':
    from keras.applications.mobilenet_v2 import MobileNetV2 as PTModel
else:
    raise ValueError('Unknown model: {}'.format(BASE_MODEL))
```

```python
input_tensor = Input(shape=(input_shape))

if transfer_learning == False:
    base_model = PTModel(include_top=False,input_tensor=input_tensor)
else:
    base_model = PTModel(weights='imagenet', include_top=False,input_tensor=input_tensor)
```

```python
last = base_model.output
x = Flatten()(last)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
```

```python
prediction = Dense(num_class, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=prediction)

if transfer_learning == True:
    for layer in base_model.layers:
        layer.trainable = False
    print("Transfer learnig with", BASE_MODEL)
```

```python
if optim=='SGD':
    from keras.optimizers import SGD
    optimizer = SGD(learning_rate=1e-4, momentum=0.9)
elif optim=='Adagrad':
    from keras.optimizers import Adagrad
    optimizer = Adagrad(learning_rate=1e-4)
elif optim=='RMSprop':
    from keras.optimizers import RMSprop
    optimizer = RMSprop(learning_rate=1e-4, rho=0.99)
elif optim=='Adadelta':
    from keras.optimizers import Adadelta
    optimizer = Adadelta(rhom=0.95)
elif optim=='Adam':
    from keras.optimizers import Adam
    optimizer = Adam(learning_rate=1e-4, beta_1=0.9, beta_2=0.999)
elif optim=='AMSGrad': 
    from keras.optimizers import Adam
    optimizer = Adam(learning_rate=1e-4, beta_1=0.9, beta_2=0.999, amsgrad=True)
else:
    raise ValueError('Unknown optimizer: {}'.format(optim))
```

```python
if num_class > 2:
    loss = "categorical_crossentropy"
else:
    loss = "binary_crossentropy"
```

```python
model.compile(loss=loss,
              optimizer=optimizer,
              metrics=['accuracy'])
```

```python
earlystop = EarlyStopping(patience=patience)

learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', # tensorflow 2.x
                                            verbose=1, 
                                            factor=1/num_class, # 1/ class数
                                            min_lr=0.00001)

modelCheckpoint = ModelCheckpoint(filepath = ".",
                                  monitor='val_accuracy',
                                  verbose=1,
                                  save_best_only=True,
                                  save_weights_only=False, # 重みのみ保存
                                  mode='max', # val_accuracyの場合
                                  save_freq=1)

callbacks = []
if early_stop:
    callbacks += [earlystop]
    
if reduce_rl:
    callbacks += [learning_rate_reduction]
    
if checkpoint:
    callbacks += [modelCheckpoint]

if len(callbacks) == 0:
    callbacks = None
callbacks
```

```python
if aug:
    history = model.fit(
        datagen.flow(x_train, y_train, batch_size=batch_size),
        steps_per_epoch=steps_per_epoch,
        epochs=epochs,
        validation_data=(x_test, y_test),
        shuffle=True,
        callbacks=callbacks)
else:
    history = model.fit(
        x_train, y_train,
        batch_size=batch_size,
        initial_epoch = 0,
        epochs=epochs,
        validation_data=(x_test, y_test),
        shuffle=True,
        callbacks=callbacks)
```

```python
scores = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])

### Plot accuracy & loss
import matplotlib.pyplot as plt 

acc = history.history["accuracy"]
val_acc = history.history["val_accuracy"]
loss = history.history["loss"]
val_loss = history.history["val_loss"]
epochs = range(1, len(acc) + 1)

#plot accuracy
plt.plot(epochs, acc, label = "Training acc" )
plt.plot(epochs, val_acc, label = "Validation acc")
plt.title("Training and Validation accuracy")
plt.legend()
plt.show()

#plot loss
plt.plot(epochs, loss,  label = "Training loss" )
plt.plot(epochs, val_loss, label = "Validation loss")
plt.title("Training and Validation loss")
plt.legend()
plt.show()
```

```python
K.set_learning_phase(1) #set learning phase

def Grad_Cam(input_model, pic_array, layer_name):
  X = np.expand_dims(pic_array, axis=0)
  preprocessed_input = X.astype('float32') / 255.0    

  grad_model = models.Model([input_model.inputs], [input_model.get_layer('block5_conv3').output, input_model.output])

  with tf.GradientTape() as tape:
    conv_outputs, predictions = grad_model(preprocessed_input)
    class_idx = np.argmax(predictions[0])
    loss = predictions[:, class_idx]

  output = conv_outputs[0]
  grads = tape.gradient(loss, conv_outputs)[0]

  gate_f = tf.cast(output > 0, 'float32')
  gate_r = tf.cast(grads > 0, 'float32')

  guided_grads = gate_f * gate_r * grads

  # 重みを平均化して、レイヤーの出力に乗じる
  weights = np.mean(guided_grads, axis=(0, 1))
  cam = np.dot(output, weights)

  # 画像化してヒートマップにして合成
  cam = cv2.resize(cam, (224, 224), cv2.INTER_LINEAR) 
  cam = np.maximum(cam, 0) 
  cam = cam / cam.max()

  jetcam = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)  # モノクロ画像に疑似的に色をつける
  jetcam = cv2.cvtColor(jetcam, cv2.COLOR_BGR2RGB)  # 色をRGBに変換
  jetcam = (np.float32(jetcam) + pic / 2)   # もとの画像に合成
  return jetcam, cam
```
